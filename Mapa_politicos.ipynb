# Importando as bibliotecas

```python
import pandas as pd # Dados do csv
import folium # Construir mapas
import geopandas as gpd # Dados do geojson
```

# Lendo o csv, e o geojson

```python
df = pd.read_csv('deputados.csv')
```

```python
gf = gpd.read_file('goias.json')
```

# Trocando o nome da coluna para fazer o join

```python
df = df.rename(columns={"CIDADE": "name"})
```

```python
gf1 = pd.concat([gf, df])
```

```python
gf1 = gf.join(df.set_index('name'), on='name')
```

# Limpando o Dataset

```python
gf1 = gf1.dropna()
```

# Criação do Mapa

```python
m = folium.Map(location=[-15.763048872811966, -47.870631102672284], zoom_start=8)
```

## Montando o Choropleth

```python
folium.Choropleth(
    geo_data = gf1,
    name='choropleth',
    data=gf1,
    columns=['name', 'VOTOS'],
    key_on='feature.properties.name',
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Aprovação da População'
).add_to(m)
```

## Montando os tooltips com o geojson

```python
folium.features.GeoJson(
    gf1,
    style_function=lambda x: {'fillColor': '#ffffff', 
                            'color':'#000000', 
                            'fillOpacity': 0.1, 
                            'weight': 0.1}, 
    control=False,
    highlight_function=lambda feature: {
        "fillColor": (
            "green" if "e" in feature["properties"]["name"].lower() else "#ffff00"
        ),
    }, 
    tooltip=folium.features.GeoJsonTooltip(
        fields=['name', 'NOME','PARTIDO','VOTOS'],
        aliases=['Cidade: ','Deputado: ','Partido: ' ,'Número de Votos: '],
        style=("background-color: white; color: #333333; font-family: arial; font-size: 12px; padding: 10px;") 
    )
).add_to(m)
```
